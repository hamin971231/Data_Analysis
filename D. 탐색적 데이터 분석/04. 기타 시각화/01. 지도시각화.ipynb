## #02. 지도 표현하기
### 1) 지도 객체 생성
지도의 중심이 되는 위도와 경도를 설정



```python
# zoom start : 배율 1~22
## 
map_osm = folium.Map(location=[37.5557098, 126.9399282],zoom_start=17)
map_osm
```

### 2) 지도객체에 마커 추가
#### 일반 마커 
아이콘 색상값 종류 


'lightgreen', 'darkgreen', 'darkblue', 'cadetblue', 'orange', 'lightred', 'darkred', 'green', 'blue', 'black', 'lightblue', 'white', 'lightgray', 'red', 'pink', 'beige', 'gray', 'purple', 'darkpurple'

```python
# 새로운 지도 객체 생성
map_osm1 = folium.Map(location=[37.5557098, 126.9399282],zoom_start=17)

## 마커객체 생성
# -> 마커가 표시될 위치 [위도,경도], 클릭시 표시될 메세지 아이콘 색상
marker1= folium.Marker([37.5557098, 126.9399282],
                        popup='메가스터디IT아카데미 신촌',
                        icon=folium.Icon(color='darked'))
marker1.add_to(map_osm1)
map_osm1
```

### 사용자 지정 아이콘, HTML 팝업

```python
### 새로운 지도 객체 생성
map_osm2 = folium.Map(location=[37.5507563, 126.9254901], zoom_start=17)

## HTML을 사용한 팝업
popup_html = folium.Popup("<div style='white-space: nowrap'><h3>홍익대학교 서울캠퍼스</h3><img src='https://megaitacademy.com/html/images//it_brand/sec08_img03.jpg' width='100%'><br/><p>tel: <a href='tel:010-7284-2384'>010-7284-2384</a></p><p>서울특별시 마포구 와우산로 94</p></div>", parse_html=False)

## 사용자 지정 아이콘 이미지 사용
## -> 온라인 상의 url에서 가져와서 사용가능 + 컴퓨터 내 경로를 이용해서 사용가능
icon_img = folium.features.CustomIcon("https://data.hossam.kr/ho.png",icon_size=(50,50))
## 마커객체 생성

marker2 = folium.Marker([37.5507563, 126.9254901],
                        popup=popup_html,
                        icon=icon_img)

## 웹페이지 파일이 저장될 폴더 생성

if not path.exists("output"):
    mkdir('output')

marker2.add_to(map_osm2)
map_osm2.save("output/map_osm2.html")
map_osm2
```

### 원형 마커(범위지정)


```python
map_osm3 = folium.Map(location=[37.5507563, 126.9254901],zoom_start=17)


# 마커 객체 생성
marker1 = folium.Marker([37.5507563, 126.9254901],  
                        icon=folium.Icon(color='orange'))

## 원형 마커
marker3 = folium.CircleMarker([37.5507563, 126.9254901],
                                radius = 100,
                                color ='#3186cc',
                                fill_color='#3186cc' )

## 원형마커 위에 아이콘 마커 올리기 
marker3.add_to(map_osm3)
marker1.add_to(map_osm3)

map_osm3
```

## #03. 서울의 고등학교 분포 확인
### 1) 데이터 준비하기

전국 초,중,고 학교위치 엑셀파일을 원격으로 내려받아 데이터 프레임으로 변환


```python
xlsx = pd.read_excel("https://data.hossam.kr/D04/school.xlsx")
xlsx.head()
```

### 2) 데이터 전처리

#### 사용할 필드만 추출 

```python
df = xlsx.filter(['학교명', '학교급구분', '소재지도로명주소', '위도', '경도'])
df.head()
```

### 서울시의 고등학교만 추출
학교급구분 필드값이 고등학교이고 소재지 도로명 주소에 서울이라는 단어가 포함된 경우


칼럼이름.str.contains("검색어")

```python
df2=df[(df['학교급구분']=='고등학교') & (df['소재지도로명주소'].str.contains("서울"))]

df2
```

### 3) 데이터 시각화

```python

map_osm=folium.Map(location=[37.566651, 126.978428],zoom_start=12)
## 글자수 초록색, 텍스트 줄바꿈 방지
html = "<font color='green' style='white-space: nowrap'><b>%s</b></font>"

## 데이터 프레임의 행 수 만큼 마커 생성
# folium.Popup("<div style='white-space: nowrap'><h3>홍익대학교 서울캠퍼스</h3><img src='https://megaitacademy.com/html/images//it_brand/sec08_img03.jpg' width='100%'><br/><p>tel: <a href='tel:010-7284-2384'>010-7284-2384</a></p><p>서울특별시 마포구 와우산로 94</p></div>", parse_html=False)
for i in df2.index :
    ## 행 우선 접근 방식으로 값 추출
    name = df2.loc[i,'학교명']
    lat = df2.loc[i,'위도']
    ing = df2.loc[i,'경도']

    ## 추출한 정보를 지도에 표시
    popup_html = folium.Popup(html % name, parse_html=False)
    marker = folium.Marker([lat,ing],popup=popup_html)
    marker.add_to(map_osm)

if not path.exists('output'):
    mkdir('output')

# 저장과 지도 출력을 함께 사용할 경우 "저장 -> 출력"순으로 사용해야 함.
map_osm.save('output/서울의_초등학교_분포.html')
map_osm
```
