```python
# 모듈 import 
import requests
import json
from pandas import DataFrame


## REST API 키값 	1dac6609b89162d4b62fa47385304f6b
## URL 값 https://dapi.kakao.com/v3/search/book
## 쿼리파라미터
## https://dapi.kakao.com/v3/search/book?query=검색어 query라는 파리미터 쓴 예

```

```python
## 접속할 url

urmlFmt="https://dapi.kakao.com/v3/search/book?query={query}&page={page}&size={size}"
```

```python
## 요청변수

query="파이썬"
page=1
size=50
key="1dac6609b89162d4b62fa47385304f6b"
```

```python
## 접속객체 생성

session=requests.Session()

## 접속객체에 header 정보 삽입

session.headers.update({
    "Authorization": "KakaoAK %s" % key
})
```

```python
## 접속할 url

url=urmlFmt.format(query=query, page=page, size=size)
url

## 생성한 접속객체를 활용해서 api 접속

r=session.get(url)

## 접속에 실패한 경우
if r.status_code!=200:
    # 에러코드와 에러 메세지 출력
    msg="{%d Error} %s 에러가 발생함" % (r.status_code,r.reason)
    # 에러를 강제로 생성시킴
    raise Exception(msg)
```

```python
## 인코딩 형식 지정

encoding='utf-8'
# 가져온 문자열을 딕셔너리로 반환

data=json.loads(r.text)
data
```

```python
## 결과를 엑셀 저장 및 출력
df=DataFrame(data['documents'])
df.to_excel("카카오 책 검색 결과(%s).xlsx" % query)
```
