# 데이터 정제
데이터 분석에 앞서 전처리가 완료된 데이터에 대해 빈값이나 이상치들을 제거하거나 다른 값으로 대체하는 처리

## #01. 결측치

- 비어있는 밧
- 실제 데이터는 수집 과정에서 발생한 오류로 인해 결측치를 포함하고 있는 경우가 많다
- 결측치가 있으면 통계 처리 함수 적용x, 분석결과가 왜곡되는 문제가 생김
- 결측값 자체의 의미가 있는 경우도 있음. --> ex) 특정 거래 자체가 존재하지 않는 경우, 인구통계학에서 아주 부자거나 아주 가난한 경우
- 결측값 처리는 작업속도에 많은 영향을 준다.
### 1. 결측치 처리 방법

#### 1) 소거법 
결측치가 포함된 행이나 열을 삭제

결측치의 비율이 미미하다면 가장 손쉬운 방법이지만, 결측치가 중요한 위치를 차지하거나 비율이 높다면 결측치 소거로 인해 통계 결과가 의미 없어질수 있다.

#### 2) 대치법
##### 평균 대치법
- 관측 또는 실험을 통해 얻어진 데이터의 ** 평균** 으로 대치
- 비조건부 평균 대치법 : 평균으로 대치
- 조건부 평균 대치법 : 회귀분석을 활용한 대치법

##### 다중 대치법

- 단순 대치법을 한번 적용하지 않고 m번의 대치를 통해 m개의 가상적 완전 자료를 만드는 방법
- 1단계 : 대치, 2단계 : 분석 3단계 : 결합

## #02. 이상치
정상 범주에서 크게 벗어난 값

### 1. 이상치로 판단하는 경우
이상값을 꼭 제거해야하는 것은 아니기 떄문에 분석의 목적이나 종류에 따라 분석가의 적절한 판단이 필요하다.

- 의도하지 않게 잘못 입력한 경우
- 의도한바와 같이입력되었으나 분석 목적에 부합되지않아 제거해야하는 경우
- 의도하지 않은 현장이지만 분석에 포함해야하는 경우
- 의도된 이상값인 경우

### 2. 극단치

이상치의 한 종류

오류는 아니지만 굉장히 드물게 발생하는 극단적인 값

> 초등학생의 몸무게 변수에 200키로 이상이면 굉장히 드문경우임 -> 극단치

### 3. 이상치의 인식과 처리

먼저 어디까지를 정상 범위로 볼 것인가를 분석가의 주관에 따라 결정해야한다.

#### 논리적으로 판단하여 정하기
성인의 몸무게가 40~150을 벗어나는 경우는 드물것이라고 판단하고 이 범위를 벅서나면 극단치로 간주

#### 통계적인 기준을 이용
상하위 0.3% 또는 '+- 3' 표전편차에 해당할만큼 극단적으로 크거나 작으면 극단치로 간주

#### 상자그림

- 중심에서 크게 벗어난 값을 극단치로 간주
- 극단치가 원으로 표시된다
- Q3,Q1을 통해 IQR을 구해 극단치 탐지
