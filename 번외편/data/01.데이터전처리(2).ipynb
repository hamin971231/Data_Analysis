### 5. 인구데이터
#### 1) 코로나 이전 배후지 직장 인구 데이터
데이터 확인

```python
df직장인구2018 = pd.read_csv('서울시 우리마을가게 상권분석서비스(상권배후지-직장인구)-2018.csv', encoding='euc-kr')
df직장인구2019 = pd.read_csv('서울시 우리마을가게 상권분석서비스(상권배후지-직장인구)-2019.csv', encoding='euc-kr')
df직장인구2020 = pd.read_csv('서울시 우리마을가게 상권분석서비스(상권배후지-직장인구)-2020.csv', encoding='euc-kr')
df직장인구2021 = pd.read_csv('서울시 우리마을가게 상권분석서비스(상권배후지-직장인구)-2021.csv', encoding='euc-kr')
df생활인구2018 = pd.read_csv('서울시 우리마을가게 상권분석서비스(상권배후지-생활인구)-2018.csv', encoding='euc-kr')
df생활인구2019 = pd.read_csv('서울시 우리마을가게 상권분석서비스(상권배후지-생활인구)-2019.csv', encoding='euc-kr')
df생활인구2020 = pd.read_csv('서울시 우리마을가게 상권분석서비스(상권배후지-생활인구)-2020.csv', encoding='euc-kr')
df생활인구2021 = pd.read_csv('서울시 우리마을가게 상권분석서비스(상권배후지-생활인구)-2021.csv', encoding='euc-kr')
# 2.
```

```python
print(df직장인구2018.shape)
print(df직장인구2018.info())

```

```python
코로나이전_직장인구_df = pd.concat([df직장인구2018,df직장인구2019])
코로나이전_직장인구_df.reset_index(drop=True,inplace=True)
drop_index = 코로나이전_직장인구_df.loc[(코로나이전_직장인구_df['기준_년_코드']==2018)&((코로나이전_직장인구_df['기준_분기_코드']==1) | (코로나이전_직장인구_df['기준_분기_코드']==2)),:].index
코로나이전_직장인구_df.drop(drop_index,inplace=True)
코로나이전_직장인구_df= 코로나이전_직장인구_df.loc[코로나이전_직장인구_df['상권_구분_코드_명']=='골목상권',:]
코로나이전_직장인구_df.reset_index(drop=True,inplace=True)
코로나이전_직장인구_df = 코로나이전_직장인구_df.loc[:,['기준_년_코드', '기준_분기_코드', '상권_코드', '총_직장_인구_수']]
코로나이전_직장인구_df.groupby(['기준_년_코드','기준_분기_코드']).count()
```

#### 2) 코로나 이후 배후지 직장 인구 데이터 

```python
코로나이후_직장인구_df = pd.concat([df직장인구2020,df직장인구2021])
코로나이후_직장인구_df.reset_index(drop=True,inplace=True)
drop_index = 코로나이후_직장인구_df.loc[(코로나이후_직장인구_df['기준_년_코드']==2018)&((코로나이후_직장인구_df['기준_분기_코드']==1) | (코로나이후_직장인구_df['기준_분기_코드']==1)),:].index
코로나이후_직장인구_df.drop(drop_index,inplace=True)
코로나이후_직장인구_df=코로나이후_직장인구_df.loc[코로나이후_직장인구_df['상권_구분_코드_명']=='골목상권',:]
코로나이후_직장인구_df.reset_index(drop=True,inplace=True)
코로나이후_직장인구_df = 코로나이후_직장인구_df.loc[:,['기준_년_코드', '기준_분기_코드', '상권_코드', '총_직장_인구_수']]
코로나이후_직장인구_df.head()
```

#### 3) 코로나 이전 배후지 생활인구 데이터

```python
df생활인구2018.head()
```

```python
df생활인구2018.columns
```

```python
코로나이전_생활인구_df = pd.concat([df생활인구2018,df생활인구2019])
코로나이전_생활인구_df.reset_index(drop=True,inplace=True)
drop_index = 코로나이전_생활인구_df.loc[(코로나이전_생활인구_df['기준_년_코드']==2018)&((코로나이전_생활인구_df['기준_분기_코드']==1) | (코로나이전_생활인구_df['기준_분기_코드']==1)),:].index
코로나이전_생활인구_df.drop(drop_index,inplace=True)
코로나이전_생활인구_df= 코로나이전_생활인구_df.loc[코로나이전_생활인구_df['상권_구분_코드_명']=='골목상권',:]
코로나이전_생활인구_df.reset_index(drop=True,inplace=True)
코로나이전_생활인구_df = 코로나이전_생활인구_df.loc[:,['기준_년_코드', '기준_분기_코드', '상권_코드', '총_생활인구_수']]
코로나이전_생활인구_df.head()
```

#### 4) 코로나 이후 배후지 생활인구 데이터

```python
코로나이후_생활인구_df = pd.concat([df생활인구2020,df생활인구2021])
코로나이후_생활인구_df.reset_index(drop=True,inplace=True)

drop_index = 코로나이후_생활인구_df.loc[(코로나이후_생활인구_df['기준_년_코드']==2018)&((코로나이후_생활인구_df['기준_분기_코드']==1) | (코로나이후_생활인구_df['기준_분기_코드']==1)),:].index
코로나이후_생활인구_df.drop(drop_index,inplace=True)
코로나이후_생활인구_df= 코로나이후_생활인구_df.loc[코로나이후_생활인구_df['상권_구분_코드_명']=='골목상권',:]
코로나이후_생활인구_df.reset_index(drop=True,inplace=True)
코로나이후_생활인구_df = 코로나이후_생활인구_df.loc[:,['기준_년_코드', '기준_분기_코드', '상권_코드', '총_생활인구_수']]
코로나이후_생활인구_df.head()
```

### 5) 인구데이터 결합
코로나 전후 직장인구 데이터 결합

```python
직장인구 = pd.concat([코로나이전_직장인구_df,코로나이후_직장인구_df])
print(직장인구.shape)
```

```python
직장인구.head()
```

코로나 전후 생활인구 데이터 결합

```python
생활인구 = pd.concat([코로나이전_생활인구_df,코로나이후_생활인구_df])
생활인구.head()
```

인구데이터 

```python
인구데이터 = pd.merge(직장인구,생활인구,left_on=['기준_년_코드','기준_분기_코드','상권_코드'],right_on=['기준_년_코드','기준_분기_코드','상권_코드'])
인구데이터.head()
```

### 6. 최종데이터프레임 구성

(매출데이터 + 상권데이터 ) + 인구데이터

```python
df = pd.merge(상권매출df,인구데이터,left_on=['기준_년_코드','기준_분기_코드','상권_코드'],right_on=['기준_년_코드','기준_분기_코드','상권_코드'])
print(df.shape)
df.head()
```
