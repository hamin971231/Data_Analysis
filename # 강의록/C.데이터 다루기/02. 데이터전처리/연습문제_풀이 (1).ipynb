### 풀이1) 리스트를 원소로 갖는 딕셔너리 활용

딕셔너리의 key가 column 이름이 된다.

```python
data = {
    "class": ["A1", "A1", "A1", "A1", "A1", 
              "A2", "A2", "A2", "A2", "A2", "A2", 
              "A3", "A3", "A3", "A3", "A3", "A3", "A3", 
              "A4", "A4", "A4", "A4", "A4", "A4", "A4"],
    "time": [62, 60, 63, 59, 61,
             62, 67, 71, 64, 65, 66, 
             68, 66, 71, 67, 68, 69, 70, 
             58, 62, 60, 59, 61, 63, 64]
}

df = DataFrame(data)
df
```

### 풀이2) 딕셔너리를 원소로 갖는 리스트 활용

```python
data = [
    {"class": "A1", "time": 62},
    {"class": "A1", "time": 60},
    {"class": "A1", "time": 63},
    {"class": "A1", "time": 59},
    {"class": "A1", "time": 61},
    {"class": "A2", "time": 62},
    {"class": "A2", "time": 67},
    {"class": "A2", "time": 71},
    {"class": "A2", "time": 64},
    {"class": "A2", "time": 65},
    {"class": "A2", "time": 66},
    {"class": "A3", "time": 68},
    {"class": "A3", "time": 66},
    {"class": "A3", "time": 71},
    {"class": "A3", "time": 67},
    {"class": "A3", "time": 68},
    {"class": "A3", "time": 69},
    {"class": "A3", "time": 70},
    {"class": "A4", "time": 58},
    {"class": "A4", "time": 62},
    {"class": "A4", "time": 60},
    {"class": "A4", "time": 59},
    {"class": "A4", "time": 61},
    {"class": "A4", "time": 63},
    {"class": "A4", "time": 64}
]

df = DataFrame(data)
df
```

### `class`를 범주형으로 변환

생성된 데이터 프레임에 대한 각 변수의 데이터 타입 확인

```python
df.dtypes
```

데이터프레임의 여러 컬럼을 일괄 변경하여 새로운 데이터프레임으로 생성

```python
df2 = df.astype({'class': 'category'})
df2.dtypes
```

특정 컬럼의 타입만 변경 후 원본에 바로 적용

```python
df['class'] = df['class'].astype('category')
df.dtypes
```

서열척도로 변경해야 하는 경우

```python
from pandas import CategoricalDtype
df['class'] = df['class'].astype(
        CategoricalDtype(categories=['A1', 'A2', 'A3', 'A4'], ordered=True))
df.dtypes
```
