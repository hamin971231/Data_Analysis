```python
# 모듈 import 
import requests
import json
from pandas import DataFrame
import os

## REST API 키값 	1dac6609b89162d4b62fa47385304f6b
## URL 값 https://dapi.kakao.com/v2/search/image
## 쿼리파라미터
## https://dapi.kakao.com/v2/search/imagek?query=검색어 query라는 파리미터 쓴 예

```

```python
## 접속할 url

urlFmt="https://dapi.kakao.com/v2/search/image?query={query}&page={page}&size={size}"
## 요청변수

query="봄 패션"
page=1
size=80
key="1dac6609b89162d4b62fa47385304f6b"

## 접속객체 생성

session=requests.Session()

#브라우저 버전정보
userAgent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"

## 접속객체에 header 정보 삽입

session.headers.update({
    "Authorization": "KakaoAK %s" % key
})
## 다음 페이지가 존재하는지 확인하기 위해 isEnd값 설정필요 : false 면 계속반복

isEnd = False

## 1페이지 부터 시작
page = 1

## 저장된 이미지 count
count=0
# 접속할 url
while not isEnd :
    url=urlFmt.format(query=query, page=page, size=size)
    ## 생성한 접속객체를 활용해서 api 접속
    r = session.get(url)
    # 접속에 실패한 경우
    if r.status_code != 200:
        msg = "[%d Error] %s 에러가 발생함" % (r.status_code,r.reason)
        raise Exception
    # 인코딩 형식 지정
    r.encoding='utf-8'
    # 가져온 문자열을 딕셔너리로 저장
    data = json.loads(r.text)
    # print(data)
    # 검색결과를 갖고있는 딕셔너리를 포함하는 리스트 추출
    searchList = data['documents']
    # print(searchList)
    ## 증감식에 해당하는 isEnd변수 추출
    isEnd=bool(data['meta']['is_end'])
    # print("is_end =",isEnd, "/ 다음페이지 = ",page)

    ## 이미지가 저장될 폴더 만들기
    if os.path.exists(query) == False :
        os.mkdir(query)
    ## 이미지 저장.
    for i,v in enumerate(searchList) :
        count  +=1
        ## 이미지를 저장하려면 이미지가 있는 url에 다시 접속해야함
        ## 이전에 머물러야하는 주소
        docUrl = v['doc_url']
        ## 이미지 url 
        imageUrl = v["image_url"]
        # print(imageUrl)
        # 다시 접속 ㄱㄱ
        # 접속객체에 header정보 삽입
        session.headers.update({
            "Authorization": "KakaoAK %s" % key,
            "Referer" : docUrl,
            "User-Agent" : userAgent
        })
        ## 텍스트 이외의 형식 가져올 떈 stream = True
        try :
            r = session.get(imageUrl,stream = True)
        except :
            continue

        # 다운로드에 실패한 경우 넘어가기
        if r.status_code !=200 :
            continue
        fileName = "%s/%05d.png" % (query,count)
        with open(fileName,'wb') as f  :
            f.write(r.raw.read())
            print(fileName," 저장 완료")
```
