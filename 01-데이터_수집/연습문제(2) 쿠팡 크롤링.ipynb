####  #01.패키지 참조

```python
# 필요한 모듈 참조
import requests
from bs4 import BeautifulSoup
from pandas import DataFrame
```

#### #02.웹 페이지 코드 수집


```python
## 수집할 컨텐츠가 있는 웹페이지 주소
url = "https://www.coupang.com/np/search?q=%EB%85%B8%ED%8A%B8%EB%B6%81&listSize=72&channel=user"

#브라우저 버전정보
userAgent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"

## 접속객체 생성
session=requests.Session()

# 접속객체에 header정보 삽입
session.headers.update({
    "Referer": "",
    "User-Agent": userAgent,
    "Accept-Language": "ko-KR,ko;q=0.8,en-US;q=0.5,en;q=0.3"
})

## 접속객체를 통해 api접속
r= session.get(url)

# 접속에 실패했을 때 

if r.status_code!=200 :
    # 200이 아니면 에러코드와 에러메세지 출력
    msg = "[%d Error] %s 에러가 발생함" % (r.status_code,r.reason)
    # 에러를 강제로 생성시킨다.
    raise Exception(msg)
    
# 인코딩 형식 지정
r.encoding = 'utf-8'
## 전체 html파일
soup = BeautifulSoup(r.text)
print(soup)
```

#### #03. 필요한 내용 추출하기

```python

productList = soup.select(".search-product-wrap")
productList 
```

```python
## 수집한 결과가 저장될 빈 리스트 
result=[]

## 제목
## 할인율
## 별점
## 리뷰수
## 원가
## 할인가



for product in productList :
    ## 상품명 추출
    nameEl = product.select(".name")
    # print(nameEl)
    name = nameEl[0].text.strip()
    ## 할인율 추출
    discountRateEl = product.select(".instant-discount-rate")
    
    if len(discountRateEl) == 0 :
        discountRateEl = 'None'
    if discountRateEl == 'None':
        discountRate = '할인없음'
    else :
        discountRate = discountRateEl[0].text.strip()
    ## 별점 추출 
    rateEl = product.select(".rating")
    # print(rateEl)
    if len(rateEl) == 0 :
        rateEl = "None"
    if rateEl == "None":
        rate = "별점없음"
    else :
        rate = rateEl[0].text.strip()
    # print(rate)
    
    ##리뷰수 추출
    rateCountEl = product.select(".rating-total-count")
    # print(rateCountEl)
    if len(rateCountEl) == 0 :
        rateCountEl = "None"
    if rateCountEl == "None":
        rateCount = "리뷰없음"
    else :
        rateCount = rateCountEl[0].text.strip()
    # print(rateCount) 
    ## 정가 추출 - el
    basePriceEl = product.select(".base-price")
    if len(basePriceEl) == 0 :
        basePriceEl = "None"
    if basePriceEl == "None" :
        basePrice = "None"
    else :
        basePrice = basePriceEl[0].text.strip()
    # print(basePriceEl)
    # ## 할인가 - el
    priceValueEl = product.select(".price-value")
    if len(priceValueEl) == 0 :
        priceValueEl= 'None'
    if priceValueEl== 'None':
        priceValue= 'None'
    else :
        priceValue = priceValueEl[0].text.strip()
    # print(priceValueEl)
    
## 상품명
## 할인율
## 별점
## 리뷰수
## 원가
## 할인가
        item = {"상품명":name,"할인율":discountRate,"별점":rate,"리뷰수":rateCount,"원가":basePrice,"할인가":priceValue}
    
    ## 딕셔너러를 준비한 리스틍 ㅔ추가 
    result.append(item)

result
```

```python
df = DataFrame(result)
df
```

```python
## 수집할 컨텐츠가 있는 웹페이지 주소
url = "https://blog.hossam.kr/"

#브라우저 버전정보
userAgent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"

## 접속객체 생성
session=requests.Session()

# 접속객체에 header정보 삽입
session.headers.update({
    "Referer": "",
    "User-Agent": userAgent    
})

## 접속객체를 통해 api접속
r= session.get(url)

# 접속에 실패했을 때 

if r.status_code!=200 :
    # 200이 아니면 에러코드와 에러메세지 출력
    msg = "[%d Error] %s 에러가 발생함" % {r.status_code,r.reason}
    # 에러를 강제로 생성시킨다.
    raise Exception(msg)
    
# 인코딩 형식 지정
r.encoding = 'utf-8'
## soup 객체 만들기
soup = BeautifulSoup(r.text)
soup
```

```python
## 필요한 내용 추출하기

articleList = soup.select(".post")
articleList
```

```python
## 수집한 결과가 저장될 빈 리스트
result=[]

## 제목, 작성일, 요약글 

for article in articleList :
    ## 제목추출
    titleEl=article.select(".entry-title > a")
    # print(titleEl)
    title=titleEl[0].text.strip()
    # print(title)
    ## 날짜 추출
    dateEl = article.select(".published")
    # print(dateEl)
    ## attrs[]는 태그안에 값을 밖으로 끄집어 내고 싶을 때
    date=dateEl[0].attrs['datetime']
    # print(date)
    # 요약글 추출
    contentEl = article.select(".entry-content > p")
    # print(contentEl)
    content = contentEl[0].text.strip()
    # print(content)

    ## 추출한 내용을 딕셔너리로
    item = {"제목":title,"날짜":date,"요약글":content}
    # print(item)

    result.append(item)
print(result)
## 데이터 프레임으로 
df = DataFrame(result)
print(df)

```
