## 01. 작업준비
### 패키지 준비


```python
import pandas as pd 
from sklearn.linear_model import LinearRegression

```

### 데이터 가져오기

```python
origin= pd.read_excel('https://data.hossam.kr/E04/cars.xlsx')
origin.head()
```

## #02. 지도학습 수행
### 분석 알고리즘을 포함하고 있는 객체 생성


```python
model = LinearRegression()
```

```python
origin[['dist']]
```

### 분석모델 학습
fit(독립변수, 종속변수)

직접 데이터를 파라미터로 전달 할 경우 2차원 배열로 설정해야함

아니면 완전한 데이터 프레임 형태로 전달해야함.

```python
fit = model.fit(origin[['speed']],origin[['dist']])
```

### 학습결과 확인 & 회귀식 

```python
## 직선의 기울기 

coef = fit.coef_
## 절편 

intercept = fit.intercept_

print ('기울기: ',coef )

print('절편 : ',intercept )

print("y = {0:.2f} * X + {1:.2f}".format(coef[0][0],intercept[0]))
```

### 설명력


```python
rscore = fit.score(origin[['speed']],origin[['dist']])
print('설명력 : ',rscore)
```

### 회귀분석 결과 활용

#### 모델을 활용한 결과값 얻기

```python
speed = [[10],[15],[20],[25],[30],[35],[40]]
pred = fit.predict(speed)
pred
```
