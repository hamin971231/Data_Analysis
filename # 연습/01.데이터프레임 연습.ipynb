# T1-1. 이상치를 찾아라(IQR활용) Expected Questions

데이터에서 IQR을 활용해 Fare컬럼의 이상치를 찾고, 이상치 데이터의 여성 수를 구하시오



```python
import pandas as pd 
import numpy as np
df = pd.read_csv("titanic.csv")
df.head()
```

```python
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR=Q3-Q1

outdata1 = df[(df['Fare'] < (Q1-1.5*IQR)) & (df['Sex']=="female")]
outdata2 = df[(df['Fare'] > (Q3+1.5*IQR)) & (df['Sex']=="female")]

print(len(outdata1)+len(outdata2))


```

# T1-4. 왜도와 첨도 구하기 (로그스케일) Expected Questions

주어진 데이터 중 train.csv에서 'SalePrice'컬럼의 왜도와 첨도를 구한 값과, 'SalePrice'컬럼을 스케일링(log1p)로 변환한 이후 왜도와 첨도를 구해 모두 더한 다음 소수점 2째자리까지 출력하시오

```python
import pandas as pd
import numpy as np

df = pd.read_csv("train (1).csv")
df.head()
```

```python
## 왜도 첨도구하기

sk = df['SalePrice'].skew()
ku = df['SalePrice'].kurt()

## 스케일링

df["SalePrice"]= np.log1p(df['SalePrice'])

sk2 = df['SalePrice'].skew()
ku2 = df['SalePrice'].kurt()

print(np.round(sk+ku+sk2+ku2,2))
```

# T1-10. 여-존슨과 박스-칵스 변환

주어진 데이터에서 20세 이상인 데이터를 추출하고 'f1'컬럼을 결측치를 최빈값으로 채운 후, f1 컬럼의 여-존슨과 박스콕스 변환 값을 구하고, 두 값의 차이를 절대값으로 구한다음 모두 더해 소수점 둘째 자리까지 출력(반올림)하시오

```python
import pandas as pd
import numpy as np

df = pd.read_csv("basic1.csv")
df.head()
```

```python
## 20세 이상 추출
df=df[df['age']>=20]
df.head()
## 최빈값 추출
## 시리즈로 출력되기 떄문에 값만 인덱싱해야함
f1_mode=df['f1'].mode()[0]

## 결측치 처리
df['f1']=df['f1'].fillna(f1_mode)
df['f1']

## 둘다 표준화가 디폴트기 때문에 standardize는 false 로 해야함
## 여-존슨 변환, 
from sklearn.preprocessing import power_transform
df['yeo'] = power_transform(df[['f1']],standardize=False)

## 박스콕스 변환
from sklearn.preprocessing import power_transform
df['box']=power_transform(df[['f1']],method="box-cox",standardize=False)
df
diff = abs(df['yeo']-df['box'])

answer = round(sum(diff),2)
answer
```

# T1-15. 슬라이싱 & 조건 Expected Question

주어진 데이터 셋에서 age컬럼 상위 20개의 데이터를 구한 다음 

f1의 결측치를 중앙값으로 채운다.

 그리고 f4가 ISFJ와 f5가 20 이상인 

 f1의 평균값을 출력하시오!


```python
import pandas as pd
import numpy as np

df = pd.read_csv("basic1.csv")
df.head()

```

```python
df=df.sort_values('age',ascending=False)[:20]
medi_f1 = df['f1'].median()
df['f1'] = df['f1'].fillna(medi_f1)
df.isnull().sum()

answer = df[(df['f4']=="ISFJ") & (df['f5']>=20)]['f1'].mean()
answer
```

# T1-16. 분산 Expected Question

주어진 데이터 셋에서 f2가 0값인 데이터를 age를 기준으로 오름차순 정렬하고

 앞에서 부터 20개의 데이터를 추출한 후 

 f1 결측치(최소값)를 채우기 전과 후의 분산 차이를 계산하시오 (소수점 둘째 자리까지)


```python
import pandas as pd
import numpy as np

df = pd.read_csv("basic1.csv")
df.head()

new_df = df[df['f2']==0].sort_values('age',ascending=True)[:20]

new_df

var_before = new_df['f1'].var()
var_before
# ## 결측치 채우기 
min_value = new_df['f1'].min()
min_value
new_df['f1'] = new_df['f1'].fillna(min_value)

new_df.isnull().sum()

var_after = new_df['f1'].var()
var_after
round((var_before-var_after),2)
```
