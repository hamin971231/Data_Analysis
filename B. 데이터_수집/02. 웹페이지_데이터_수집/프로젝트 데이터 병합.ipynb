```python
# 필요한 모듈 참조
import pandas as pd
from pandas import DataFrame    
from pandas import read_excel
```

```python
## 미즈노 상품정보가 담긴 데이터프레임 불러오기
df_mizuno = read_excel("미즈노_상품정보.xlsx")
df_mizuno.head()
```

```python
## 아디다스 상품정보가 담긴 데이터프레임 불러오기
df_adidas = read_excel("아디다스_상품정보.xlsx")
df_adidas.head()
```

```python
## 두 데이터프레임 병합
df_final=pd.concat([df_mizuno,df_adidas])
df_final
```

```python
## 데이터 프레임의 필드 이름이 한글이면 에러 -> 영어로 바꾸는 과정
df_convert = df_final.rename(columns={"상품이름": "name", "정가": "price", 
                    "최종가격 ": "sale_price", "리뷰수": "review", "이미지링크": "img"})

df_convert
```
