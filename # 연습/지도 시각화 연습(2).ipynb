```python
df = pd.read_csv('제주관광공사_콘텐츠.csv',encoding='cp949')
df
```

```python
df.columns
```

```python
# 쓸만한 컬럼만 추출
con_2022 = df[["콘텐츠분류", "제목", "언어", "지번주소", "도로명주소", "위도", "경도", "평점", "조회수", "좋아요수", 
                  "리뷰수", "북마크수", "소셜네트워크공유수", "인기점수"]].copy()
# 결측치 확인
con_2022.isnull().sum()
```

```python
# 제목 결측치 확인

con_2022[con_2022['제목'].isnull()]
```

```python
# 평점 NAN 삭제
con_2022 = con_2022.dropna(axis=0)
```

```python
con_2022.isnull().sum()
```

```python
con_2022['언어'].unique()
```

```python
kor_con_2022 = con_2022.loc[con_2022['언어']=='국문',:].reset_index(drop=True)
kor_con_2022.head()
```

```python
kor_con_2022['평점'].value_counts().sort_index(ascending=False)
```

```python
food = kor_con_2022[kor_con_2022['콘텐츠분류']=='음식점']
food = food[food['평점']==5.0]
len(food)
```

```python
top_food = food.sort_values('인기점수',ascending=False)
top_food.head()
```

```python
len(top_food)
```

```python
top_food.iloc[3,5]
```

```python
import folium

# 중심지정
jeju_map = folium.Map(location=[33.3616666, 126.5291666],zoom_start=11)

for i in range(0,len(top_food),1):
    folium.Marker([top_food.iloc[i,5],top_food.iloc[i,6]],popup=top_food.iloc[i,1],icon=folium.Icon(color='blue')).add_to(jeju_map)

jeju_map
```
