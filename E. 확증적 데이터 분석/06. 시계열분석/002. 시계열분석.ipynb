### 기본형태의 시계열 그래프

```python
plt.figure()
sb.lineplot(df,x=df.index,y='age')
plt.grid()
plt.show()
plt.close()
```

### 평균 이동
주어진 구간의 평균을 구해서 이어서 표시하는 방법

평균 이동 결과가 실제 데이터와 얼마나 차이가 나는지의 여부로 데이터를 판단

#### 1) 단순 이동평균 (SMA, Simple Moving Average)
가장 많이 사용하는 방식
3년, 6년, 9년단위 평균

```python
df['sma3'] = df['age'].rolling(3).mean()
df['sma6'] = df['age'].rolling(6).mean()
df['sma9'] = df['age'].rolling(9).mean()
df
```

```python
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize = (15,10))

sb.lineplot(df,x=df.index,y='age',ax=ax1)
sb.lineplot(df,x=df.index,y=df['sma3'],ax=ax2)
sb.lineplot(df,x=df.index,y=df['sma3'],ax=ax3)
sb.lineplot(df,x=df.index,y=df['sma3'],ax=ax4)
```

> 평균이동 시각화의 목적은 데이터가 정상성을 만족한느지 확인하여 차분 적용여부를 결정하기 위함임.
