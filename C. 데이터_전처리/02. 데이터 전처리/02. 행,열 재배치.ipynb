# 데이터 전처리 (1)  - 행,열 재배치

## #01. 패키지 참조


```python

from pandas import read_excel
import numpy

from pandas import Series
```

### 데이터 가져오기

```python
df = read_excel('https://data.hossam.kr/pre_sample.xlsx',index_col='이름')

df
```

### 데이터 확인

```python
## 결측치 채우기
df = df.fillna(96)

## 데이터타입 바꾸는 함수 astype
## 원본이 바뀌지 않기 때문에 df = df.astype~ 이런식으로 가야함. 
df = df.astype({'국어' : int,'과학': int,"수학":str})
df.dtypes
```

## #02. 행 열 재배치 이름변경

### 1. 열 순서 변경
    - 지정된 수ㅜㄴ서대로 열이 재정렬된 결과가 반환된다. 
    - 원본은 변화 없음. 결과가 적용된 복사본이 생성된다.
    - 재배치 되지 못한 열은 재회된다. 

```python
## 칼럼의 순서를 재배치 
## 재배치 안하면 빠짐 

## 지정된 수선서대로 열이 재정렬된 값 반환
## 원본은 변화 없음
df1= df.reindex(columns=['국어','수학','과학','영어'])
df1
```

### 행의 순서 변경

    - 행의 순서를 재배치 
    - 재배치 안하면 빠짐 
    - 지정된 수선서대로 행이 재정렬된 값 반환
    - 원본은 변화 없음

```python

df2= df.reindex(index=['석영','용식','수현','민철'])
df2
```

### 행이나 열의 이름 변경

- 기존이름 : 새이름 형식의 딕셔너리로 지정
- columns 와 index중 변경을 원하는 하나만 설정가능
- 원본은 변화 없음
- 이름이 변경되지 않은 항목은 제외되지 않는다
- 운본에 즉시 반영해야 할 경우에는 inplace = True 파라미터 적용


```python
df3 = df.rename(columns= {"국어":'kor','수학':'math'},
                index = {"수현":'suhyeon','용식':'yong'},
                inplace = False)


df3

## 원본에 반영시키고 싶으면 inplace = True
```

## #06. 파생변수 추가

기존 변수들의 조합, 변형, 또흔 계산을 통해 얻어지는 새롱누 변수


### 1. 데이터 새로 가져오기 

```python
import pandas as pd
```

```python
df= pd.read_excel("https://data.hossam.kr/C02/pre_sample.xlsx",index_col='이름')
```

#### 1) 칼럼간의 연산

```python
df1=df.copy()

df1['직접합계']=df['국어'] + df['영어'] + df['수학'] + df['과학']
df1
```

### 2. 기술 통계 값관련 함수를 사용한 파생변수

합계, 평균, 최대, 최소, 중앙값, 분산, 표준편차


```python
df2=df.copy()

df2['합계'] = df2[df2.columns[2:6]].sum(axis=1)
df2
```

```python
# 0번째 행만 가져오기
df.iloc[0]

## 1~4번째 전까지 행만 가져오기

df.iloc[1:4]
```

```python
## 모든 행에서 3번째 열만 가져옴

df.iloc[:,3]
```

```python
## 모든 행에 대해서 2번째부터 끝까지 열을 추출하여 합계
df.iloc[:,2:6].sum(axis=1)

```

```python
# iloc : index-location (인덱스번호 기반) / loc : location (이름기반)
# 평균
df.loc[:, '국어':].mean(axis=1)
```

```python
# 최대
df.loc[:, '국어':].max(axis=1)
```

```python
# 최소
df.loc[:, '국어':].min(axis=1)
```

```python
# 표준편차
df.loc[:, '국어':].std(axis=1)
```

```python
# 분산
df.loc[:,'국어':].var(axis=1)
```

```python
df5=df.copy()
df5['합계'] = df5.loc[:,"국어":].sum(axis=1)
df5['평균'] = df5.loc[:,"국어":"과학"].mean(axis=1)
df5
```

#### 평균값에 따른 조건부 파생변수

```python
import numpy as np
```

```python
df5['결과'] = np.where(df5['평균']>=80 ,'합격','불합격')
df5
```

#### 경우의 수에 따른 조건값

```python
# 학점을 부여하기 위한 점수의 구간을 설정하는 조건들을 리스트로 설정
conditions = [  (df5['평균'] >= 90),   # A
                (df5['평균'] >= 80),   # B
                (df5['평균'] >= 70)    # C
]

value = ["A","B","C"]

df5['학점'] = np.select(conditions, value,default="F")
df5
```
