## #01. 주성분 분석 개요
### 차원 축소(Dimesionality Reduction)
데이터의 전반적인 특성을 보존하면서 데이터의 변수 수를 줄이는 방법

> 크기가 10행 7열 dataframe의 차원은? (10*7)

> 변수 3개를 제거하면 10행 4열로 바뀜. 이떄의 차원은 (10*4)

|방법|조오류|
|--|--|
|특성 선택|	가장 중요한 특성들만 선택하여 기존의 데이터를 표현|
|특성 추출|	기존 특성들을 사용하여 새로운 특성들을 만들어내는 방법|

### 주성분 분석(PCA)
데이터의 가장 큰 분산을 가진 방향으로 차원을 축소하여 데이터를 표현하는 방법

이를 통해 데이터를 가장 잘 설명하는 주요 특성들을 찾을 수 있다.

데이터의 복잡성을 줄여주어 다양한 분야에서 활용되며, 머신러닝, 패턴 인식, 시각화, 데이터 압축 등 다양한 분야에서 중요한 기술로 사용

```python
df = pd.read_excel("https://data.hossam.kr/E04/boston.xlsx")
df.drop('CAT. MEDV', axis=1, inplace=True)
df.head()
```

## #02. 데이터 전처리
### 독립변수 칼럼만 추출


```python
x_train = df.drop("MEDV",axis=1)
x_train
```

### 추출된 독립변수를 표준화

```python
scaler = StandardScaler()
x_train_std = scaler.fit_transform(x_train)
x_train_std
```

### 표준화 결과를 데이터 프레임으로 재구성


```python
std_df = pd.DataFrame(x_train_std,columns=x_train.columns)
std_df
```

## #02.Sklearn을 사용한 PCA 분석

```python
# 주성분 분석 객체 생성 (n_components: 주성분 개수)

model = PCA(n_components=5)
fit = model.fit_transform(std_df)
fit
```

### 결과 확인

sklearn의 pca분석은 기존의 변수들을 토대로 n_component 수 만큼 새로운 변수를 생성한다

대부분 머신러닝의 학습데이터를 생성하는 용도


```python
pca_df = pd.DataFrame(fit)
pca_df
```

## #03.PCA 패키지를 사용한 분석 (추천)


```python
# 주성분 분석의 대상 칼럼수를 독립변수의 전체 칼럼의 길이로 설정

model = pca(n_components=len(std_df.columns))

# 표준화 결과를 활용하여 주성분 분석 수행
fit = model.fit_transform(std_df)
fit
```

### 생성된 주성분에 사용된 필드 확인

```python
fit
```

```python
len(std_df.columns)
```

```python
topfit = fit['topfeat']
topfit
```

## #04. 주성분 분석 결과를 토대로 회귀분석 수행

### 원본 데이터 프레임으로 분석
