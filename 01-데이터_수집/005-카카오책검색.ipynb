```python
# 모듈 import 
import requests
import json
import datetime as dt
from pandas import DataFrame
```

```python
# 온라인 상의 url 받아오기
urlFmt = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key={key}&targetDt={targetDt}"
```

```python
# 발급받은 API 연동키
API_KEY = "6d2cf4aa96725383235c717f2e569f1e"
```

```python
# 하루 전 날짜 얻기 

date = dt.datetime.now() - dt.timedelta(days=1) ## now에서 하루 뺼거임.
yesterday= date.strftime('%Y%m%d')
```

```python
# 최종 url
url = urlFmt.format(key=API_KEY, targetDt=yesterday)
# 어제 날짜를 조회하겠다는 것을 url에 입력하는것.
```

```python
# 접속 객체 생성

session = requests.Session()

## 접속 정보 설정 
session.headers.update({
    ## 직전에 머물렀던 페이지
    'Referer' : "",
    ## 브라우저 및 os정보 
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
})
```

```python
## 특정 웹페이지 접속

r = session.get(url)
```

```python
# 접속에 실패 한 경우

if r.status_code!=200 :
    # 에러코드와 에러 메세지 출력
    msg="[%d Error] %s 에러가 발생함." %{r.status_code, r.reason}
    # 에러를 강제로 생성시킴
    raise Exception(msg)

```

```python
# 인코딩 형식 지정

r.encoding='utf-8'

# 가져온 문자열을 딕셔너리로 전환 

data=json.loads(r.text)
print(data)
```

```python
# 필요한 부분만 추출

dailyBoxOfficeList=data['boxOfficeResult']['dailyBoxOfficeList']
print(dailyBoxOfficeList)
```

```python
# 결과 표 생성하기

df=DataFrame(dailyBoxOfficeList)
df
```

```python
# 엑셀로 저장 
df.to_excel("박스오피스_순위_%s.xlsx" % yesterday)
```
