### 문제2
다음은 상장된 금융사 중 외국기업(F)과 내국기업(K)의 주가이다. 외국기업과 내국기업별로 기술통계량을 상자그림을 그려서 비교/분석하라. 최종적으로 분석 결과를 토대로 하여 알 수 있는 사실을 하나 이상 제시하라.


```python
import pandas as pd 
from matplotlib import pyplot as plt
import numpy as np
import datetime as dt
import seaborn as sb



# 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'   # 윈도우 전용
#plt.rcParams['font.family'] = 'AppleGothic'    # 맥 전용
plt.rcParams["font.size"] = 10

# 그래프 크기 설정
plt.rcParams["figure.figsize"] = (12,8)

# 유니코드에서 음수 부호설정
plt.rcParams['axes.unicode_minus'] = False
df = pd.read_excel("https://data.hossam.kr/D02/stock.xlsx")
df.head()
```

```python
## 외국기업 기술 통계량

foreign = df[df['구분']=='F']
foreign_desc = foreign.describe()
foreign_desc



```

```python
## 국내기업 기술 통계량

korean = df[df['구분']=='K']
korean_desc = korean.describe()
korean_desc


```

```python
## 박스상자 

fig, ax = plt.subplots(1,2,figsize=(20,15))
plt.figure()

## 전체 제목
fig.suptitle("외국기업과 국내기업 상자그림")

## 외국기업 박스그림
sb.boxplot(data=foreign,y='주가',ax=ax[0])
ax[0].set_ylim(100,250)
ax[0].set_title("외국기업 주가")
ax[0].grid()

## 국내기업 박스그림

sb.boxplot(data=korean,y='주가',ax=ax[1])
ax[1].set_ylim(100,250)
ax[1].set_title("국내기업 주가")
ax[1].grid()


```

####  - 외국기업의 주가가 국내기업의 주가보다 높지만 주가의 중앙값은 국내기업이 더 높다
####  - 상자그림을 그려본 결과 국내기업의 주가와 외국기업의 주가에 각각 하나의 이상차 존재
--------------------------------------------------------------
#### - 주가의 변동 폭은 국내기업이 외국기업보다 더 크다
#### - 국내기업은 폭락지점, 외국기업은 폭등지점이 보인다

### 문제3

포도농장에서 수집한 비료 종류에 따른 수확량 자료이다. 기술통계량을 구하고 상자그림을 그려라. 최종적으로 분석 결과를 토대로 하여 알 수 있는 사실을 하나 이상 제시하라.

```python
import pandas as pd 
from matplotlib import pyplot as plt
import numpy as np
import datetime as dt
import seaborn as sb



# 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'   # 윈도우 전용
#plt.rcParams['font.family'] = 'AppleGothic'    # 맥 전용
plt.rcParams["font.size"] = 10

# 그래프 크기 설정
plt.rcParams["figure.figsize"] = (12,8)

# 유니코드에서 음수 부호설정
plt.rcParams['axes.unicode_minus'] = False
df = pd.read_excel("https://data.hossam.kr/D02/grape.xlsx")


print(df[df['비료종류']=="A"]['수확량'])
df
```

```python
## 비료 종류별로 기술통계량 확인
df.groupby("비료종류").agg(["min",'max','mean','median','var'])

```

```python
## 박스상자 

fig, ax = plt.subplots(1,2,figsize=(20,15))
plt.figure()

## 전체 제목
fig.suptitle("비료별 수확량 상자그림")

## A 박스그림
sb.boxplot(data=df,y=df[df['비료종류']=="A"]['수확량'],ax=ax[0])
ax[0].set_ylim(10,40)
ax[0].set_title("A 비료 수확량")
ax[0].grid()

## B 박스그림

sb.boxplot(data=df,y=df[df['비료종류']=="B"]['수확량'],ax=ax[1])
ax[1].set_ylim(10,40)
ax[1].set_title("B 비료 수확량")
ax[1].grid()
plt.show()
plt.close()
```

#####  - 수확량의 최소값, 최대값 , 평균, 중앙값을 확인한 결과 수확량은 A사료가 더 좋다
#####  - 하지만 분산과 박스 그림을 확인한 결과 A에서 2개의 이상치와 A의 분산이 B에비해 높은것을 확인했다

### 문제4

다음은 통계학개론을 수강하는 두 학과(A, C)의 성적분포이다. 이 두 그룹을 비교하기 위한 기술통계량을 구하고 상자그림을 그려서 비교/분석하라. 최종적으로 분석 결과를 토대로 하여 알 수 있는 사실을 하나 이상 제시하라

```python
import pandas as pd 
from matplotlib import pyplot as plt
import numpy as np
import datetime as dt
import seaborn as sb



# 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'   # 윈도우 전용
#plt.rcParams['font.family'] = 'AppleGothic'    # 맥 전용
plt.rcParams["font.size"] = 10

# 그래프 크기 설정
plt.rcParams["figure.figsize"] = (12,8)

# 유니코드에서 음수 부호설정
plt.rcParams['axes.unicode_minus'] = False
df = pd.read_excel("https://data.hossam.kr/D02/analysis_grade.xlsx")

df.head(10)
```

```python
## groupby 써서 min max mean, median var 

df.groupby('학과').agg(['min','max','mean','median','var'])

```

```python
df_A = df[df['학과']=="A"]

df_C = df[df['학과']=="C"]
## 박스상자 

fig, ax = plt.subplots(1,2,figsize=(20,15))
plt.figure()

## 전체 제목
fig.suptitle("학과 별 상자그림")

## A 박스그림
sb.boxplot(data=df_A,y='점수',ax=ax[0])
ax[0].set_ylim(0,100)
ax[0].set_title("A학과 점수")
ax[0].grid()

## C 박스그림

sb.boxplot(data=df_C,y='점수',ax=ax[1])
ax[1].set_ylim(0,100)
ax[1].set_title("C학과 점수")
ax[1].grid()
plt.show()
plt.close()
```

##### - A학과는 C학과에 비해 점수의 범위가 크고, 분산이 C학과에 비해 큰것으로 보아 데이터가 C학과에 비해 산재되어있을것이다.
##### - A학과보다 C학과의 최소값, 최댓값, 평균값, 중앙값이 더 높은것을 보아 성적이 전체적으로 우수할 것이다.
##### - 박스 그림을 확인한 결과, C학과의 성적은 평균주변으로 몰려있고 특별히 못하거나 잘하는 이상치가 존재

##### (1) 전공에서 STAT를 1로, COMP를 2로 변환한 레이블을 적용하라.



```python
import pandas as pd 
from matplotlib import pyplot as plt
import numpy as np
import datetime as dt
import seaborn as sb



# 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'   # 윈도우 전용
#plt.rcParams['font.family'] = 'AppleGothic'    # 맥 전용
plt.rcParams["font.size"] = 10

# 그래프 크기 설정
plt.rcParams["figure.figsize"] = (12,8)

# 유니코드에서 음수 부호설정
plt.rcParams['axes.unicode_minus'] = False
df = pd.read_excel("https://data.hossam.kr/D02/stat_comp_grade.xlsx")

df.head(10)


```

```python
df[df['전공']=="STAT"] = df[df['전공']=="STAT"].replace("STAT","1")
df[df['전공']=="COMP"] = df[df['전공']=="COMP"].replace("COMP","2")

df 
# df['전공']=np.where(df['전공']=='STAT',1,2)
```
