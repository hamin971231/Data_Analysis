## #01. 작업준비

## #03. 추론통계
### 1) 상관분석

```python
rdf = pearson_r(df)
```

```python
rdf
```

상관분석 결과 p값이 0.05보다 작으므로 속도와 제동거리간 상관관계가 유의하다고 할 수 있다.

### 2) 회귀분석
$y=ax+b$ 에 따라 종속 ~ 독립 형태로 표현식 구성

#### 분석을 위한 학습모델 구성



```python
model = ols("dist ~ speed",data=df)
model
```

#### 분석 수행


```python
fit = model.fit()
fit
```

#### 분석결과 확인 

```python
fit.summary()
```

#### 분석결과 해석

##### R-squared(R제곱, 결정계수)

모형의 설명력을 의미 : dist 제동거리를 speed가 약 65% 설명한다

##### Adj R-squared
독립변수가 여러 개인 다중회귀분석에서 사용 

독립변수의 개수와 표본의 크기를 고려하여 R-squared 값 보정

서로다른 모형을 비교할 떄는 이 지표가 높은 쪽을 선택한다 

##### F-statistic: 89.57, Prob(F-statistic): 1.49e-12
회귀모형에 대한 (통계적) 유의미성 검증 결과, 유의미함 (p < 0.05)

즉, 이 모형은 주어진 표본 뿐 아니라 모집단에서도 의미있는 모형이라 할 수 있음

##### AIC, BIC
로그 우도: 종속변수가 정규분포라고 가정했을 때의 우도

로그우도를 독립변수의 수로 보정한 값 (작을 수록 좋다)

##### 회귀계수 해석

데이터로부터 얻은 계수의 추정치 

절편의 추정치는 -17.5791

speed 의 계수 추정치는 3.9324로 speed가 1 증가할 떄마다 dist가 3.9324 증가한다는 것으ㅡㄹ 의미

-> dist = -17.5791+ 3.9324*speed

##### p값
모집단에서 계수가 0일 때, 현재와 같은 크기의 표본에서 이러한 계수가 추정될 확률

이 확률이 매우 작다는 것은, 모집단에서 speed의 계수가 정확히 3.9324는 아니더라도 현재의 표본과 비슷하게 0보다 큰 어떤 범위에 있을 가능성이 높다는 것을 의미

보통 5%와 같은 유의수준을 정하여 p값이 그보다 작으면(p < 0.05), "통계적으로 유의미하다"라고 정리



## #04. 결과 보고
### 모형적합도 보고
F분포의 파라미터 2개와 그 때의 F값, p-value의 유의수준 비교를 명시

dist에 대하여 speed로 예측하는 회귀분석을 실시한 결과, 이 회귀모형은 통계적으로 유의미하였다(F(1,48) = 89.57, p < 0.05)
### 독립변수에 대해 보고
speed의 회귀계수는 3.9324로, dist에 대하여 유의미한 예측변인인 것으로 나타났다.(t(48) = 9.464, p < 0.05)
