# 코로나 확진자 시계열 분석


## #01. 작업준비

### 패키지 참조

```python
import sys
import os
sys.path.append(os.path.dirname(os.path.dirname(os.getcwd())))


from datetime import datetime as dt
from datetime import timedelta
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import dates as mdates
from statsmodels.tsa.arima.model import ARIMA
from pmdarima.arima import auto_arima
import seaborn as sb
from helper import set_datetime_index, exp_timedata
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
```

```python
origin = pd.read_excel("https://data.hossam.kr/E06/covid19_seoul_230531.xlsx")
origin.head()
```

## #02. 데이터 전처리

### 필요한 데이터만 추출

```python
df = origin.filter(['서울시 기준일', '서울시 추가 확진'])
df.rename(columns={'서울시 기준일':'date', '서울시 추가 확진':'confirmed'}, inplace=True)
df.head()
```

### 각 필드의 데이터 타입 확인

외부에서 가져온 데이터는 항상 원하는 타입인지 확인 후 필요하다면 타입 변환을 거쳐야한다.

```python
df.dtypes
```

### 날짜 타입에 대한 형변환

```python
df['date'] = pd.to_datetime(df['date'].str.strip(), format='%Y-%m-%d')
df.dtypes
```

### 결측치 검사


```python
df.isnull().sum()
```

### 결측치 정제

결측치인 경우는 확진자 발생하지 않은것으로 간주하고 0 으로 치환

```python
df2 = df.fillna(0)
df2.isnull().sum()
```

### date 필드를 날짜 형식의 인덱스로 저장

```python
df3 = set_datetime_index(df2,'date')
df3.head()
```

## #03. 데이터 검정
이상치는 보이지만 데이터 자체가 실제 발생한 현상이므로 정상 데이터로 판단함



```python
exp_timedata(data=df3,yname='confirmed',sd_model='a')
```

## #04. ARIMA 분석

분석모델 만들기 


```python
model = ARIMA(df3['confirmed'],order = (1,0,0),seasonal_order=(1,0,0,7))
fit = model.fit()
print(fit.summary())
```

### 학습데이터에 대한 예측치
학습한 데이터에 대한 predict함수 결과값을 내장하고 있다 . 


```python
fv =fit.fittedvalues
fv.head()
```

### 학습한 내용을 토대로 이후 120일 간의 예상치 생성


```python
fc = fit.forecast(120)
fc.head()
```

### 시각화

```python
last = df3.index.max()
xmin = last - timedelta(days=120)
xmax = last + timedelta(days=120+10)
ymax = df3['confirmed'][xmin:xmax].max()
```

```python
plt.figure(figsize=(20,8))

## 원본 데이터 
sb.lineplot(data=df3,x=df3.index , y='confirmed',label = 'Original')
## 원본에 대한 학습결과
sb.lineplot(x=fv.index,y=fv.values,label = 'FittedValues',linestyle = '--')
## 향후 120일 간의 예측치 
sb.lineplot(x=fc.index,y=fc.values,label = 'Predict', linestyle='--',color = 'red')

plt.xlabel('Day')
plt.ylabel('Confirmed')
plt.legend()
plt.xlim([xmin, xmax])

plt.ylim([0, ymax*1.2])

## 그래프의 x축이 날짜로 구성되어 있을 경우 형식 지정
monthyearfmt = mdates.DateFormatter('%y.%m.%d')
plt.gca().xaxis.set_major_formatter(monthyearfmt)
plt.grid()
plt.show()
plt.close()
```

## #04. Auto Arima 분석

### 분석 수행

```python
my_p = 1    # AR의 차수 (검증한 결과를 활용)
my_d = 0    # 차분 횟수 (검증한 결과를 활용)
my_q = 0    # MA의 차수 (검증한 결과를 활용)
my_s = 7    # 계절성 주기 (분석가가 판단)

model = auto_arima(
    y=df3['confirmed'],  # 모델링하려는 시계열 데이터 또는 배열
    start_p=0,           # p의 시작점
    max_p=my_p,          # p의 최대값
    d=my_d,              # 차분 횟수
    start_q=0,           # q의 시작점
    max_q=my_q,          # q의 최대값
    seasonal=True,       # 계절성 사용 여부
    m=my_s,              # 계절성 주기
    start_P=0,           # P의 시작점
    max_P=my_p,          # P의 최대값
    D=my_d,              # 계절성 차분 횟수
    start_Q=0,           # Q의 시작점
    max_Q=my_q,          # Q의 최대값
    trace=True           # 학습 과정 표시 여부
)
print(model.summary())
```

### 학습한 데이터와 동일 기간에 대한 예측치 산정

```python
fv = model.fittedvalues()
fv.head()
```

### 향후 120일 간의 예측치

```python
fc = model.predict(n_periods=120)
fc.head()
```

### 시각화

```python
last = df3.index.max()
xmin = last - timedelta(days=120)
xmax = last + timedelta(days=120+10)
ymax = df3['confirmed'][xmin:xmax].max()
```

```python
plt.figure(figsize=(20,8))

# 원본데이터
sb.lineplot(data=df3,x=df3.index,y='confirmed',label = 'Original')

## 원본에 대한 학습결과
# 원본에 대한 학습결과
sb.lineplot(x=fv.index, y=fv.values, label='FittedValues', linestyle='--')

# 향후 120일간의 예측값
sb.lineplot(x=fc.index, y=fc.values, label='Predict', linestyle='--', color='red')

plt.xlabel('Day')
plt.ylabel('Confirmed')
plt.legend()

plt.xlim([xmin, xmax])
plt.ylim([0, ymax*1.2])

# 그래프의 x축이 날짜로 구성되어 있을 경우 형식 지정
monthyearFmt = mdates.DateFormatter('%y.%m.%d')
plt.gca().xaxis.set_major_formatter(monthyearFmt)

plt.grid()
plt.show()
plt.close()
```
