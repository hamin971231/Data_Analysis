## #03. 뉴런과 활성화 함수의 이해

### 1) 뉴런

- 0~1사이의 숫자
- 위 그램에서 위치한 0.4,0.3,0.9 의 조합에 따라 오른쪽 뉴런의 값이 결정된다.
- 여기서 화살표 실선의 역할이 중요
- 화살표 실선은 어떤 실수를 의미함. -> +일수도 있고, - 일수도 있다
- 이 실수를 가중치라 부른다.

#### 가중치를 임의로 결정한 경우

뉴런의 값과 가중치의 값을 곱한 결과를 모두 합한 결과에 임의의 값인 편향을 더한 결과가 ?의 값이 된다.

but 편향을 있을수도 있고, 없을수도 있다.


위의 그림에서 ?에 들어갈 결과랎은 (-2.12+0.36+0.72)+2.3 이 나오기 때문에 뉴런은 0~1사이의 숫자라는 가정에 위배된다.

그러므로 이 결과값은 어떠한 처리를 거쳐 0~1 사이 값으로 변환 되어야 한다.

### 2) '어떠한 처리' = 활성화 함수


가중치나 편향은 특별한 제약없이 모든 real_number가 될 수 있으므로 무한대에 가까운 양수나 실수가 될 수 있다.

그러므로 계산한 결과값이 0~1사이의 값이 아닐수도있다.

활성화 함수는 파라미터로 어떤 값이 전달되더라도 0~1 사이의 값으로 바꿔줌

활성화 함수는 파라미터값을 변환한 우에도 상대적인 크기가 유지되어야한다.

즉, 활성화 함수는 모든 x에 대해 우상향이다.

### 3) 정리

1. 뉴런의 값과 가중치의 값을 곱한다.
2. 곱한 값을 모두 더한다.
3. 여기에 편향을 더한다. -> 결과값 k값
4. 결과값 k를 0~1사이로 만들어주는 함수 f(x)에 전달한다

위 과정에서 k값을 axtivation이라고 하고 0~1사이의 값으로 변환해 주는 함수 f(x)를 활성화 함수라고 한다.

#### 뉴런의 계산
- 뉴런은 활성화 함수를 이용해 출력을 결정하며 입력신호의 가중치 합을 계산하여 임계값과 비교한다. 
- 가중치 합이 임계값보다 작으면 뉴런의 출력은 -1, 같거나 크면 1을 출력한다.

## #04. 활성화 함수의 종류

### 1) 선형함수
- 단층 퍼셉트론에서 사용
- 선형함수를 제외한 모든 활성화 함수는 비선형 활성화 함수임

### 2) 시그모이드 함수

- 로지스틱 회귀 분석과 유사
- 0~1 사이의 확률값을 갖는다.
- 이진분류의 출력층 노드에서 0~1 사이의 값을 만들고 싶을 때 사용
- 단점: input값이 너무 크거나 작으면 기울기가 0이 된다.

### 3) 하이퍼볼릭 탄젠트 함수 

- 시그모이드 함수의 크기와 위치를 조절한 함수.
- 범위는 -1 ~ 1 
- 그래프의 모양은 0을 기준으로 대칭
- 이 때문에 하이퍼볼릭탄젠트는 시그모이드 활성화 함수로 썼을 때 보다 학습 수렴속도가 빠름. 
### 4) 소프트맥스 함수

- 표준화지수 함수로도 불림
- 출력값이 여러 개로 주어지고 목표치가 다범주인 경우 각 범주에 속할 확률을 제공하는 함수.
- 카테고리 분류에 사용

### 5) 렐루 함수

- 렐루 함수는 입력이 양수인 경우에는 입력값을 그대로 출력하고, 음수인 경우에는 0을 출력
### 6) 손실함수 이해

머신러닝, 딥러닝은 결국 컴퓨터가 독립변수로 부터 종속변수에 도달하기 위한 가중치와 편향을 찾는 과정

- 손실 함수(Loss Function)은 머신 러닝 및 딥 러닝 모델에서 모델의 예측값과 실제 타겟(목표) 값 사이의 차이를 측정하는 함수
- 손실 함수는 모델이 얼마나 정확한 예측을 하는지를 평가하고, 이 평가를 기반으로 모델의 가중치(weight) 및 매개변수(parameter)를 조정하는 데 사용
- 손실 함수의 목표는 모델의 예측값과 실제 타겟 값 사이의 차이를 최소화하는 것

### 경사하강법 
- 1차 방정식의 근삿값을 발견하기 위한 최적화 알고리즘
- 신경망은 가중치를 업데이트 하면서 주어진 문제를 최적화
- 경사하강법은 특정 함수에서의 기울기를 이용하여 최적의 값을 찾아 손실함수를 최소화 하는 방법

### 손실함수의 종류

#### 회귀분석
- MSE, MAE
#### 이진 분류
- 교차 엔트로피 손실(Cross-Entropy Loss)
#### 다중분류
- 범주형 교차 엔트로피 (Sprase-Categorical-Cross-Entropy-Loss)
