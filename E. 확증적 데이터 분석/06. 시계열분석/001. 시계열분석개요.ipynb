# 시계열 분석 개요
## #01. 시계열 자료의 이해

- 시간의 흐름에 따라 관찰된 값들 
- 시계열 데이터의 분석을 통해 미래의 값을 예측하고ㅡ 주기, 계절성 등을 파악하여 활용한다.

### 1. 시계열 자료의 종류

|종류|설명|
|--|--|
|비정상성 시계열 자료| 시계열 분석을 실시할 때 다루기 어려운 자료로 대부부의 시계열 자료|
|정장 시계열 자료 | 비정상 시계열을 핸들링해 다루기 쉬운 시계열 자료로 변환한 자료|


#### 정상성
> 평균과 분산이 일정하고 공분산도 단지 시차에만 의존하고 특정 시점에는 의존하지 않는 상태 

#####  평균이 일정할 경우

- 모든 시점에 대해 일정한 평균을 갖는다. 
- 실제 대부분의 자료는 평균이 일정하지 않다. 이 경우 차분(difference)을 통해 정상화 할 수 있다.
##### 분산이 일정할 경우
- 분산도 특정 시점에 의존하지 않고 일정해야 한다.
- 분산이 일정하지 않을 경우 변환을 통해 변환(transformation)을 통해 정상화 할 수 있다. 

##### 약한 의미의 정상성 (약정상성)
모든 시점에 평균 일정, 시점과 분산 독립, 공분산은 시차에만 의존 
##### 비정상 -> 정상: 변환 & 차분
|이름|설명|
|--|--|
| 변환 | 분산이 일정하지 않은 비정상 시계열에 대해 수행|
|차분|평균이 일정하지 않은 비정상 시계열에 대해 수행|
|일반차분 | 바로 전 시점의 자료를 빼는 방법이다 .|
|계정차분 | 여러 시점 전의 자료를 빼는 방법, 주로 계절성을 갖는 자료를 정상화 하는데 사용|

## #02. 시계열 분석 모형
### 1. AR 모형(자기회귀모형)
- 특정 시점 전의 자료가 현재 자료에 영향을 주는 형태
- 판단 조건 : 자기 상관 함수(ACF)가 빠르게 감소하고 부분 자기 상관 함수(PACF)는 어느 시점에 절단점을 가짐.

### 2. MA 모형(이동평균모형)
- 유한한 개수의 백색잡음의 결합
- 항상 정상성을 만족 
- ACF에서 잘단점을 갖고, PCAF가 빠르게 감소
### 3. ARIMA 모형(자기회귀누적이동평균모형)
- 가장 일반적인 모형
- 비정상시계열 모형
- 차분이나 변환을 통해 AR/MA/ARMA 모형으로 정상화 가능
- 세가지 지표를 사용하는 p,d,q모형이라고도 함

|종류|의미|
|--|--|
|$p$|AR과 관련이 있는 차수|
|$q$|MA모형과 관련있는 차수|

$p$와 $q$는 ARIMA모형에서 ARMA로 정상화 할 때 차분한 횟수를 의미
|종류|의미|
|--|--|
|$p = 0$|IMA(d,q) 모형이라고 부르고 , d번 차분하면 MA(q)모형을 따른다|
|$d = 0$|ARMA(d,q) 모형이라고 부르고 , 정상성을 만족한다.|
|$q = 0$|ARI(p,d) 모형이라고 부르고, d번차분하면 AR(p)모형을 따른다. |

#### 예시
|예|내용|
|--|--|
|$ARIMA(0,1,1)$|1차분 후 $MA(1)$모형을 활용 |
|$ARIMA(1,1,0)$|1차분 후 $AR(1)$모형을 활용 |

## #03. 분해 시계열
- 시계열에 영향을 주는 일반적ㅇ니 요인을 시계열에서 분리해 분석하는 방법
- 회귀분석적 방법 주로 사용
### 시계열 구설ㅇ요소
|요소|설명 |
|--|--|
|추세요인|자료가 어떤 특정한 형태를 취할 때|
|계절요인|고정된 주기에 따라 자료 변화|
|순환요인(cyclical factor)	|알려지지 않은 주기를 갖고 자료가 변화|
|불규칙요인(irregular factor)|회귀분석에서 오차에 해당하는 요인|
