```python
### 지도 이미지를 다른 것으로 사용하고 싶으면 tiles 옵션 사용 ㄱㄱ
m = folium.Map(location=[37.53897093698831, 127.05461953077439],
zoom_ttart = 14,
tiles = 'Stamen Terrain')
m
```

```python
m = folium.Map(location=[37.53897093698831, 127.05461953077439],
zoom_start = 14)
# 뚝섬역
folium.Marker([37.54706945947954, 127.04740975332888]).add_to(m)
# 성수역
folium.Marker([37.54461957910074, 127.05590699103249],
        popup='<b>subway</b>').add_to(m)

## 건대입구역 
folium.Marker([37.54041716624373, 127.06914637466906],
        popup='<b>건대입구역</b>',
        tooltip='<i>핫플</i>').add_to(m)

m
```

```python
## folium.clickForMarker() 지도위에 마우스를 클릭하면 마커를 생성해준다.

m = folium.Map(location=[37.53897093698831, 127.05461953077439],
        zoom_start=14)

m.add_child(folium.ClickForMarker(popup='마커'))
m
```

```python
## 위 , 경도를 지도에서 확인하고 싶을때 LatLngPopup()

m = folium.Map(location=[37.53897093698831, 127.05461953077439],
        zoom_start=14)

m.add_child(folium.LatLngPopup())
```

```python
# folium.Circle(), folium.CircleMaker()
# 원의 범위 구하기
m = folium.Map(location=[37.53897093698831, 127.05461953077439],
zoom_start = 14)

# circle

folium.Circle(
    location = [37.54455152583346, 127.05564949938811],
    radius=50,
    color='#eb9e34', # 원 선 색상
    fill_color='red', # 원 내부 색상
    popup='Circle popup',
    tooltip='Circle tooltip'
).add_to(m)

# CircleMarker
folium.CircleMarker(
    [37.54455152583346, 127.05564949938811],
    radius=100,
    fill=True,
    color='#34ebc6',
    fill_color='#c634eb',
    popup='CircleMarker popup',
    tooltip='CircleMarker tooltip'
).add_to(m)

m
```
